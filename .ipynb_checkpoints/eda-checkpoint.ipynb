```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('dataset_cleaned.csv')
```

```python
df = df.drop(columns=['id'])
```

```python
df
```

```python
df['distance_to_nearest_city'] = df['distance_to_nearest_city'] / 1000
```

```python
df
```

```python
for col in df.columns:
    if col not in ['address']:
        df[col].hist()
        plt.xlabel(col)
        
        plt.xticks(rotation=90)
        plt.ylabel('Frequency')
        plt.show()
```

2. Bivariate Analysis

```python
for val in df['nearest_city'].unique():
    sns.scatterplot(x=np.log1p(df[df['nearest_city']==val]['price_per_m2']+1), y=np.log1p(df['price']+1), data=df)
    plt.title('{} vs Price'.format(val))
    plt.show()
```

```python
df['distance_to_nearest_city'].describe()
```

```python
sns.scatterplot(x=df['bedroom_nums'], y=df['bathroom_nums'], data=df)
plt.title('Bedroom vs Bathroom')
plt.show()
```

```python
sns.scatterplot(x=df['bathroom_nums'], y=df['car_spaces'], data=df)
plt.title('Bedroom vs Bathroom')
plt.show()
```

3. Correlation matrix

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

4. Grouped aggregation

```python
df.groupby('nearest_city')['price'].mean().sort_values(ascending=False).plot(kind='bar')
plt.title('Average Price for each city')
plt.show()
```

```python
df.groupby('nearest_city')['price_per_m2'].mean().sort_values(ascending=False).plot(kind='bar')
plt.title('Average Price per m2 by city')
plt.show()
```
